```python
z=df.iloc[:,[0,1,2,3,4,5,6]].values #To extract columns with index 0,1,2,3,4,5,6
x_df = pd.DataFrame(z)
x_df
```

```python
#Determine Optimum number of cluster by elbow method
from sklearn.cluster import KMeans
plt.figure(figsize=(14,7))

wcss = [] #(Within-Cluster Sum of Square). WCSS is the sum of the squared distance between each point and
#the centroid in a cluster.
for x in range(1,11):
    kmeans = KMeans(n_clusters =x, init= 'k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(z)
    wcss.append(kmeans.inertia_)

#plot the results
plt.plot(range(1,11), wcss,color='red')
plt.title('The Elbow Method Graph',fontweight='bold',fontsize=13,bbox={'facecolor':'0.9','pad':6},pad=18)
plt.xlabel('Number of clusters(k)',fontsize=15)
plt.ylabel('WCSS',fontsize=15)
plt.grid()
plt.show()
```

```python
#training the k-means model on a dataset
kmeans = KMeans(n_clusters =4, init= 'k-means++', max_iter=300, n_init=10, random_state=0)
y_predict= kmeans.fit_predict(z)

a = df['Crops']
y_means = pd.DataFrame(y_predict)

cluster_df = pd.concat([y_means, a], axis=1)
cluster_df =cluster_df.rename(columns= {0:'cluster'})
cluster_df.reset_index(drop=True,inplace=True)
cluster_df
```

### 6: Model Selection.

Now we break our dataset into two parts:
1) Dependent variables
2) Independent Variables

```python
#axis=1 means column
a = df.drop(['Crops'], axis=1) #Independent variables
a.head()
```

```python
b = df['Crops'] #Dependent Variables
b.head()
```

```python
print('shape of a:', a.shape)
print('shape of b:', b.shape)
```

```python
from sklearn.model_selection import train_test_split
#The ideal split is said to be 80:20 for training and testing respectively.
x_train, x_test, y_train, y_test = train_test_split(a,b, test_size=0.2, random_state=0)
print("The Shape of x train:", x_train.shape)
print("The Shape of x test:", x_test.shape)
print("The Shape of y train:", y_train.shape)
print("The Shape of y test:", y_test.shape)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')
li1=[]
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
a=lr.score(x_train,y_train)
li1.append(a)
print("Accuracy Score of logistic Regression:",a)

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(x_train,y_train)
b=kn.score(x_test,y_test)
li1.append(b)
print("Accuracy Score of KNN:",b)

from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
c=svc.score(x_test,y_test)
li1.append(c)
print("Accuracy Score of Support Vector Machine:",c)

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy',random_state=7)
dtc.fit(x_train,y_train)
d=dtc.score(x_test,y_test)
li1.append(d)
print("Accuracy Score of decision tree classifier:",d)

from sklearn.naive_bayes import GaussianNB
nv=GaussianNB()
nv.fit(x_train,y_train)
e=lr.score(x_train,y_train)
li1.append(e)
print("Accuracy Score of Naive Bayesian Classifier:",e)

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
f=lr.score(x_train,y_train)
li1.append(f)
print("Accuracy Score of Random Forest Classifier:",f)
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
g=lr.score(x_train,y_train)
li1.append(g)
print("Accuracy Score of Gradient Boosting Classifier:",g)
from sklearn.neural_network import MLPClassifier
h=MLPClassifier()
h.fit(x_train,y_train)
h=lr.score(x_train,y_train)
li1.append(h)
print("Accuracy Score of Artificial Neural Network Classifier:",g)
```
